## 3. Indexing and Selection

Selecting and filtering data is one of the most common tasks in data mining. Pandas provides powerful indexing options using `.loc`, `.iloc`, slicing, and Boolean masks.



```python
# âœ… Task for students: Use .loc (label-based) to print first or second row
print(df.loc[0])
print('*****************************')
print(df.loc[1])
```

```python
# the loc works with any values used in the index regardless of the type
df.index = [f"row_{i+1}" for i in range(len(df))] #this line replaces the DataFrameâ€™s index (row labels) with a list of custom strings:like 0---> row_1
df.loc["row_2"]
```

```python
# âœ… Task for students: Select all passengers under the age of 18
from termcolor import colored
'Using iloc'
a= df.loc[df['age'] < 18]
b = df[df['age'] < 18].iloc[:]
display(b)

if a.equals(b):
    print(colored('Both iloc and loc give the same output', 'red'))
```

```python
# Create a new column for age group
import numpy as np

def tell_age(x):
    if x <= 12:
        return 'Child'
    elif x <= 17 :
        return 'Teen'
    else:
        return 'Adult'

df['age_group']=df['age'].apply(tell_age)
#df.loc[df.age_group == 'Adult']
# Calculate survival rate per group
df.groupby('age_group')['survived'].mean()    #df.groupby(['col1', 'col2'])
                                              #df.groupby('column_name')['value_column'].mean()

```
